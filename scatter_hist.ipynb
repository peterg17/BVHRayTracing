```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
idx_data = pd.read_csv("perf_data.csv")
idx_data
```

```python
ax = sns.displot(idx_data, x="wall_time", hue="indexer", multiple="stack")
sns.set(rc={"figure.facecolor": "white"})
ax.set(xlabel='wall_time (sec.)')
ax.savefig('idx_histogram.png')
```

```python
ax = sns.displot(idx_data, x="wall_time", hue="indexer", kind="kde", fill=True)
ax.set(xlabel='wall_time (sec.)')
ax.savefig('idx_curve')
```

```python
octree_data = idx_data.loc[idx_data['indexer'] == "OCTREE"]
bvh_data = idx_data.loc[idx_data['indexer'] == "BVH"]
bvh_sah_data = idx_data.loc[idx_data['indexer'] == "BVH_SAH"]
octree_data
```

```python
octree_avg = octree_data["wall_time"].mean()
octree_median = octree_data["wall_time"].median()
print("octree avg time: ", octree_avg)
print("octree median time: ", octree_median)
```

```python
bvh_avg = bvh_data["wall_time"].mean()
bvh_median = bvh_data["wall_time"].median()
print("bvh avg time: ", bvh_avg)
print("bvh median time: ", bvh_median)
```

```python
bvh_sah_avg = bvh_sah_data["wall_time"].mean()
bvh_sah_median = bvh_sah_data["wall_time"].median()
print("bvh sah avg time: ", bvh_sah_avg)
print("bvh sah median time: ", bvh_sah_median)
```
